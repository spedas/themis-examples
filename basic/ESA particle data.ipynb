## ESA particle data

This notebook demonstrates basic examples of accessing ESA particle data

```python
import pyspedas
from pytplot import tplot
```

Load all calibrated (l2) ground processed data products

```python
pyspedas.themis.esa(trange=['2010-02-13', '2010-02-14'], probe='b')
```

Plot the ion density, average temperature, velocity (GSM coordinates) and energy flux

```python
tplot(['thb_peif_density', 'thb_peif_avgtemp', 'thb_peif_velocity_gsm', 'thb_peif_en_eflux'])
```

Load all calibrated (l2) on-board data products

```python
pyspedas.themis.mom(trange=['2010-02-13', '2010-02-14'], probe='b')
```

Plot the electron density and velocity (in DSL coordinates)

```python
tplot(['thb_peem_density', 'thb_peem_velocity_dsl'])
```
