## All Sky Imager data

This notebook shows how to load the THEMIS ASI keograms

note: the MPLPlotter version of pytplot is available via:

pip install --upgrade https://github.com/MAVENSDC/PyTplot/archive/matplotlib-backend.zip

```python
from pytplot.MPLPlotter.tplot import tplot
import pyspedas
```

```python
trange = ['2008-02-10', '2008-02-11']

pyspedas.themis.ask(trange=trange)
```

Set some plot options and plot the full day

```python
from pytplot import options, tplot_options, tlimit

options('thg_ask_fykn', 'Colormap', 'gray')
options('thg_ask_inuv', 'Colormap', 'gray')
options('thg_ask_fsim', 'Colormap', 'gray')
options('thg_ask_fsmi', 'Colormap', 'gray')

options('thg_ask_fykn', 'yrange', [0, 255])
options('thg_ask_inuv', 'yrange', [0, 255])
options('thg_ask_fsim', 'yrange', [0, 255])
options('thg_ask_fsmi', 'yrange', [0, 255])

options('thg_ask_fykn', 'zrange', [0, 1e4])
options('thg_ask_inuv', 'zrange', [0, 8e3])
options('thg_ask_fsim', 'zrange', [0, 6e3])
options('thg_ask_fsmi', 'zrange', [0, 1e4])

tplot_options('title', 'THEMIS ASI Examples')

tlimit(['2008-02-10 00:00:00', '2008-02-11 00:00:00'])

tplot(['thg_ask_fykn','thg_ask_inuv','thg_ask_fsim','thg_ask_fsmi'])
```
