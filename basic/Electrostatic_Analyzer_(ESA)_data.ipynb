```python
import pyspedas
from pytplot import tplot, options
```

Load all calibrated (l2) ground processed data products

```python
trange = ['2010-02-13', '2010-02-14']
probe = 'b'

pyspedas.themis.esa(trange=trange, probe=probe)
```

Set some metadata, then plot the fast survey electron density, velocity and energy flux

```python
options('th'+probe+'_peef_en_eflux', 'ylog', True)
options('th'+probe+'_peef_en_eflux', 'zlog', True)

tplot(['th'+probe+'_peef_density', 'th'+probe+'_peef_velocity_dsl', 'th'+probe+'_peef_en_eflux'])
```
