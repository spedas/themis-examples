This notebook shows how to load data from the THEMIS EFI instrument

For information on these data products, see:

```python
import pyspedas
from pytplot import tplot
```

Load the L1 data, and append the suffix '_raw' to the variables

```python
pyspedas.themis.efi(trange=['2007-07-20', '2007-07-21'], level='l1', probe=['c', 'd'], get_support_data=True, suffix='_raw')
```

Set some plot metadata on the L1 variables

```python
from pytplot import options
options('thc_vaf_raw', 'color', ['purple', 'blue', 'teal', 'green', 'yellow', 'red'])
options('thd_vaf_raw', 'color', ['purple', 'blue', 'teal', 'green', 'yellow', 'red'])
options('thc_vaf_raw', 'legend_names', ['V1', 'V2', 'V3', 'V4', 'V5', 'V6'])
options('thd_vaf_raw', 'legend_names', ['V1', 'V2', 'V3', 'V4', 'V5', 'V6'])

options('thc_eff_raw', 'color', ['b', 'g', 'r'])
options('thd_eff_raw', 'color', ['b', 'g', 'r'])
options('thc_eff_raw', 'legend_names', ['e12', 'e34', 'e56'])
options('thd_eff_raw', 'legend_names', ['e12', 'e34', 'e56'])


tplot(['thc_vaf_raw', 'thd_vaf_raw', 'thc_eff_raw', 'thd_eff_raw'])
```

zoom in on the data

```python
from pytplot import tlimit

tlimit(['2007-07-20 17:36:00', '2007-07-20 17:45:00']) # Mozer intervals

tplot(['thc_vaf_raw', 'thd_vaf_raw', 'thc_eff_raw', 'thd_eff_raw'])
```

Update the y-axis range

```python
options('thc_eff_raw', 'yrange', [-1000, 1500])
options('thd_eff_raw', 'yrange', [-2000, 3000])

tplot(['thc_vaf_raw', 'thd_vaf_raw', 'thc_eff_raw', 'thd_eff_raw'])
```

Load the L2 data

```python
pyspedas.themis.efi(trange=['2007-07-20', '2007-07-21'], probe=['c', 'd'], level='l2')
```

```python
options('thc_eff_dot0_dsl', 'yrange', [-10, 15])
options('thc_eff_dot0_gse', 'yrange', [-15, 15])
```

```python
tplot([ 'thc_vaf_raw', 'thc_eff_raw', 'thc_eff_dot0_dsl', 'thc_eff_dot0_gse'])
```
