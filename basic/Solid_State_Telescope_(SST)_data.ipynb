```python
import pyspedas
from pyspedas import tplot, options
```

Load the L2 data

```python
trange = ['2010-06-05','2010-06-06']
probe = 'c'

pyspedas.projects.themis.sst(trange=trange, probe=probe)
```

Set some metadata and create a figure showing the SST electron energy flux, velocity, average temperature and density

```python
options('th'+probe+'_psef_en_eflux', 'ylog', True)
options('th'+probe+'_psef_en_eflux', 'zlog', True)

tplot(['th'+probe+'_psef_en_eflux',
       'th'+probe+'_psef_velocity_gse', 
       'th'+probe+'_psef_avgtemp', 
       'th'+probe+'_psef_density'])
```
